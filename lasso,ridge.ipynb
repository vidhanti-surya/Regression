```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns



import warnings
warnings.filterwarnings("ignore")
```

```python
sf=sns.load_dataset("mpg")
```

```python
sf
```

```python
sf.drop("name",axis=1,inplace=True)
```

```python
sf
```

```python
sf.head()
```

```python
sf.tail()
```

```python
sf.info()
```

```python
sf.isna().sum()
```

```python
#cheak the data type 
sf.dtypes
```

```python
#since we are not done outlieear tritmeant ,then better idea would to be replace the missing value with meddian
sf["horsepower"].median()
```

```python
#now i   have to fill the null vallue with the meddian value
sf['horsepower']=sf['horsepower'].fillna(sf["horsepower"].median())
```

```python
sf.isna().sum()#there are no missing value
```

```python
sf['origin'].unique()
```

```python
sf['origin'].value_counts()
```

```python
#this is categorical col then we have to chane the value in to num cal
#its called data encoding 
```

```python
#data encoding
sf["origin"]=sf["origin"].map({"usa":1,"japan":2,"europe":3})
```

```python
sf["origin"]
```

```python
#we have to convert the method into int but its by defult here int 
sf["origin"]=sf["origin"].astype(int)
```

```python
#seprated the data into   x and y 

```

```python
x=sf.drop("mpg",axis=1)
y= sf["mpg"]
```

```python
x
```

```python
y
```

```python
#now i have to do train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train
```

```python
#now  have to build the m0del linear regression model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model
```

```python
#train the model 
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
# i want this coef in the column size

for i ,col_name in enumerate(x_train.columns):
    print(f"the coefficient for {col_name} is {model.coef_[i]}")
```

```python
#observation :
#coefficients are relatively smaller ,if one independentvariable change  there
#will be  not much diffrence in prediction 
#this is sometimes called as smoother model


#these features might not be contributinng in model training
```

```python
#model predicition
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score_lin=r2_score(y_pred,y_test)
print(f" the square of liner regression{r2_score_lin}")
```

```python
#now i  have to regularised model
#ridge regression 
```

```python
from sklearn.linear_model import Ridge
ridge_model=Ridge(alpha =0.1)
```

```python
ridge_model
```

```python
#i  have  to train the model 
ridge_model.fit(x_train,y_train)
```

```python
# i want this coef in the column size in ridge regression

for i ,col_name in enumerate(x_train.columns):
    print(f"the coefficient for {col_name} is {ridge_model.coef_[i]}")
```

```python
#in ridge regression the coefficient never will be zero
```

```python
#for the ridge  regression  evaulation
y_pred_ridge=ridge_model.predict(x_test)
r2_score_rid=r2_score(y_pred_ridge,y_test)
print(f"the r 2 squre of ridge regression{r2_score_rid}")
```

```python
#we dont see much variation in coef of ridge regression as compared to liner regression
```

```python
lasso
```

```python
grid_searh
```
