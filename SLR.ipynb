```python
#we reach the data so -now we have to do  the EDA , DATACLEANING ,FEATURE ENGINEERING
```

```python
#AFTER FEQATURE ENGINEERING WE HAVE TO SHEPRATE THE DATA AND SELECT THE XX AND Y
x = data[["bmi"]] #one independent variable 
y = data["target"]#dependent variable
```

```python
from sklearn.model_selection import train_test_split

#give the size for the data (test_size)
#every time the data is randomly change so for fixing the data i have to fix it (random_state = 1)
#its mean the vlue is only one time change 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state=1)
```

```python
#now i ahve to do model training 
from sklearn.linear_model import LinearRegression # here the class is linerRegression
```

```python
model1 = LinearRegression() # here we made the object of the model
```

```python
model1
```

```python
# read the box here the nothing so its empaty instence 
x_train
```

```python
y_train
```

```python
# for the model training 
model1.fit(x_train,y_train)
#at the backend it will use gradient  descent to give optimal coefficient
```

#here the  color is change  here the model is fitted

```python
# now we have to do model test becz our the model testing is complete 
model1.predict(x_test)
```

```python
y_pred =model1.predict(x_test)
```

```python
y_pred
```

```python
#visualise the result 
plt.scatter(x_test,y_test,color = "black",label = "Actual")
plt.plot(x_test,y_pred,color = 'blue',linewidth = 3,label= "predicted liner line")
plt.xlabel("BMI")
plt.ylabel("one year  progresion target ")
plt.title("liner regression on diabetes data")
plt.legend()
plt.show()
```

```python
#now we have to do model evalation 
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```
