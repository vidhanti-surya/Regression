```python
#1st step?? i  have to seprate the x and y
x=df1.iloc[:,:-1] #independemt variable
y=df1.iloc[:,-1 ]#target variavble 
```

```python
#MODEL SELECTION OF THE MODEL
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state = 1)
```

```python
x_train
```

```python
y_train
```

```python
#now i have to scalling but its is the optional part but i  have to do


from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler
```

```python
#fit transfrom the data 
```

#always remamber part 
fit_transfrom data for the train data set 
but fit  for the test data set 

```python
x_train=scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

```

```python
x_train
```

```python
x_test
```

```python
#scaleing is done so i  have to builod the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model
```

```python
#for the training the model 
model.fit(x_train,y_train)
```

```python
#now the  model is reay
model.coef_
```

```python
model.intercept_
```

```python
#now i have to predict the model\
#jab bhi hmm xtest ko predicit krte hai wo mera y predict ban jata hai ussi ke upr hmm model ka price batayenge 

y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
#now  i  have to do performance metrics\
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
#now  we find the root mean square error
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
score=r2_score(y_test,y_pred)
```

```python
score
```

```python
#NOW I AM CALCULATING ADJUSTED R2
1-(1-score) *(len(y_test)-1) / (len(y_test)-x_test.shape[1]-1)
```

```python
# adjusted r2 is lower then r2 then its called good model 
```
